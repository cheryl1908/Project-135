```python
import matplotlib.pyplot as plt
import csv
import seaborn as sns

rows = []

with open("star_with_gravity.csv", "r") as f:
    reader = csv.reader(f)

    for row in reader:
        rows.append(row)

headers = rows[0]
star_data_rows = rows[1:]
```

```python
names = []
distances = []
masses = []
radiuses = []
gravity = []
```

```python
for star_data in star_data_rows:
    if len(star_data) == 6:
        names.append(star_data[1])
        distances.append(star_data[2])
        masses.append(star_data[3])
        radiuses.append(star_data[4])
        gravity.append(star_data[5])

print(names)
```

```python
plt.figure(figsize=(15,7))
plt.bar(names[0:9],masses[0:9])
plt.xlabel("Name")
plt.ylabel("Mass")
plt.title("Stars Solar Mass") 
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.bar(names[0:9],radiuses[0:9])
plt.xlabel("Name")
plt.ylabel("Radii")
plt.title("Stars Solar Radius") 
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.bar(names[0:9],distances[0:9])
plt.xlabel("Name")
plt.ylabel("Distance")
plt.title("Stars Solar Distance") 
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.bar(names[0:9],gravity[0:9])
plt.xlabel("Name")
plt.ylabel("Gravity")
plt.title("Stars Solar Gravity") 
plt.show()
```
